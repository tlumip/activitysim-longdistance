```python
# many TAZs have no possible destinations for duration 2 complete tours with at least 10 minutes in the destination
count = 0
for i in range(1, 5117):
    time = np.where(dists.loc[i] >= 50, times.loc[i], 999999).min()
    print(time)
    if time > 60:
        count += 1
```

```python
x = []
for i in range(1, 5117):
    time = np.where(dists.loc[i] >= 50, times.loc[i], 999999).min()
    if 2 * time + min_time_in_dest >= 119 and land_use.loc[i]["MODELAREA"] == 1:
        print(2 * time + min_time_in_dest)
        x.append(i)
```
